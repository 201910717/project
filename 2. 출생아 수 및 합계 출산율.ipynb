# 2. 출생아 수 및 합계 출산율


```python
import pandas as pd
df = pd.read_excel('stat_142801.xls', skiprows=2,nrows=2,index_col=0)
df
#앞에 불필요한 부분은 skiprows로 삭제하고 필요한 row 2개만 추출
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'},inplace=True)
df
```

```python
df.loc['출생아 수']
```

```python
df.iloc[0] #정수를 이용해 위치로 인덱스 불러옴
```

```python
df=df.T
```

```python
df
```

```python
import matplotlib.pyplot as plt 
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows 버전 글꼴 변경
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index,df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.plot(df.index,df['출생아 수'])

ax2 = ax1.twinx()  #x축을 공유하는 쌍둥이 axis를 만들어줌
ax2.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700) #출생아 수 y 범위 설정
ax1.set_yticks([300,400,500,600]) #그래프에 데이터 표현
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val, ha='center') #val 그래프랑 숫자 공중에 띈 정도

ax2 = ax1.twinx()  #x축을 공유하는 쌍둥이 axis를 만들어줌
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5) 
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'],color='#ffd100',marker='o',markersize = 10,markeredgecolor='w',mew=3,linewidth=5)
```
